```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets
from sklearn.decomposition import KernelPCA
from sklearn.datasets import fetch_mldata
from mpl_toolkits.mplot3d import Axes3D
```

```python
iris = datasets.load_iris()
iris_X = iris.data
iris_y = iris.target
iris_target_names = iris.target_names
```

```python
kernel = "cosine" #“linear” | “poly” | “rbf” | “sigmoid” | “cosine” | “precomputed”
```

```python
kpca = KernelPCA(n_components=2, kernel=kernel)
iris_X_r = kpca.fit_transform(iris_X)
for i, iris_target_name in zip(range(len(np.unique(iris_y))), iris_target_names):
    plt.scatter(iris_X_r[iris_y == i, 0], iris_X_r[iris_y == i, 1], alpha=.8, label=iris_target_name)
plt.legend(loc='best')
plt.title('PCA of IRIS dataset')
plt.show()
```

```python
kpca = KernelPCA(n_components=3, kernel=kernel)
iris_X_r = kpca.fit_transform(iris_X)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
for i, iris_target_name in zip(range(len(np.unique(iris_y))), iris_target_names):
    ax.scatter(iris_X_r[iris_y == i, 0], iris_X_r[iris_y == i, 1],  iris_X_r[iris_y == i, 2], alpha=.8, label=iris_target_name)
plt.legend(loc='best')
plt.title('PCA of IRIS dataset')
plt.show()
```

```python
mnist = fetch_mldata('MNIST original')
samples, _ = mnist.data.shape
selection = (np.random.rand(samples)>0.99)
mnist_X = mnist.data[selection]
mnist_y = mnist.target[selection]
```

```python
kpca = KernelPCA(n_components=2, kernel=kernel)
mnist_X_r = kpca.fit(mnist_X).transform(mnist_X)
for i in range(len(np.unique(mnist_y))):
    plt.scatter(mnist_X_r[mnist_y == i, 0], mnist_X_r[mnist_y == i, 1], alpha=.7, label=i)
plt.legend(loc='best')
plt.title('PCA of MNIST dataset')
plt.show()
```

```python
kpca = KernelPCA(n_components=3, kernel=kernel)
mnist_X_r = kpca.fit(mnist_X).transform(mnist_X)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
for i in range(len(np.unique(mnist_y))):
    ax.scatter(mnist_X_r[mnist_y == i, 0], mnist_X_r[mnist_y == i, 1], mnist_X_r[mnist_y == i, 2], alpha=.6, label=i)
plt.legend(loc='best')
plt.title('PCA of MNIST dataset')
plt.show()
```
